# Download and Prepare KuaiRec Dataset

This notebook downloads and prepares the KuaiRec dataset for our experiments.

**Dataset Source:** https://kuairec.com/

**Dataset Info:**
- Fully observed user-item interaction matrix from Kuaishou video-sharing app
- Each user has viewed every video (almost complete matrix)
- Files: big_matrix.csv (main interactions), small_matrix.csv (test subset)
- Format: user_id, video_id, watch_ratio, timestamp


```python
# Import libraries
import urllib.request
import os
import numpy as np
import pandas as pd
from pathlib import Path
import zipfile
```

## Step 1: Download and Extract Dataset

This cell automatically downloads and extracts the KuaiRec dataset.


```python
# Download and extract KuaiRec dataset using wget and unzip
import subprocess
import shutil
import os
from pathlib import Path

# Create data directory
data_dir = Path('../data/kuairec_data')
data_dir.mkdir(parents=True, exist_ok=True)
print(f"Data directory: {data_dir.absolute()}")

dataset_url = "https://nas.chongminggao.top:4430/datasets/KuaiRec.zip"
zip_path = data_dir / "KuaiRec.zip"

# Change to data directory for wget/unzip commands
original_cwd = os.getcwd()
os.chdir(data_dir)

# Download using wget
if zip_path.exists():
    print(f"✓ Zip file already exists: {zip_path}")
    print("  Skipping download. Delete the zip file if you want to re-download.")
else:
    print("Downloading KuaiRec dataset using wget...")
    print(f"URL: {dataset_url}")
    print("This may take a few minutes...\n")
    
    try:
        # Use wget with --no-check-certificate
        result = subprocess.run(
            ['wget', dataset_url, '--no-check-certificate'],
            check=True,
            capture_output=True,
            text=True
        )
        print("✓ Download complete!")
        if zip_path.exists():
            print(f"  File size: {zip_path.stat().st_size / (1024*1024):.2f} MB")
    except subprocess.CalledProcessError as e:
        print(f"✗ wget failed: {e}")
        print("  Trying Python download method...")
        import ssl
        ssl._create_default_https_context = ssl._create_unverified_context
        try:
            urllib.request.urlretrieve(dataset_url, zip_path)
            print("✓ Download complete using Python!")
        except Exception as e2:
            print(f"✗ Download failed: {e2}")
            print("  Please run manually:")
            print(f"  wget {dataset_url} --no-check-certificate")
    except FileNotFoundError:
        print("✗ wget not found. Trying Python download method...")
        import ssl
        ssl._create_default_https_context = ssl._create_unverified_context
        try:
            urllib.request.urlretrieve(dataset_url, zip_path)
            print("✓ Download complete using Python!")
        except Exception as e:
            print(f"✗ Download failed: {e}")
            print("  Please install wget or run manually:")
            print(f"  wget {dataset_url} --no-check-certificate")

# Extract using unzip
if zip_path.exists():
    big_matrix_check = data_dir / "big_matrix.csv"
    
    if big_matrix_check.exists():
        print(f"\n✓ big_matrix.csv already exists. Skipping extraction.")
        print("  Delete big_matrix.csv if you want to re-extract.")
    else:
        print(f"\nExtracting {zip_path.name} using unzip...")
        try:
            # Use unzip command
            result = subprocess.run(
                ['unzip', '-o', 'KuaiRec.zip'],
                check=True,
                capture_output=True,
                text=True
            )
            print("✓ Extraction complete!")
            
            # Move CSV files to data_dir if they're in a subdirectory
            for root, dirs, files in os.walk(data_dir):
                for file in files:
                    if file in ['big_matrix.csv', 'small_matrix.csv']:
                        src = Path(root) / file
                        dst = data_dir / file
                        if src != dst and not dst.exists():
                            shutil.move(str(src), str(dst))
                            print(f"  ✓ Moved {file} to data directory")
                            
        except subprocess.CalledProcessError as e:
            print(f"✗ unzip failed: {e}")
            print("  Trying Python zipfile method...")
            try:
                with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                    zip_ref.extractall(data_dir)
                    print("✓ Extraction complete using Python!")
            except Exception as e2:
                print(f"✗ Extraction failed: {e2}")
                print("  Please run manually:")
                print("  unzip KuaiRec.zip")
        except FileNotFoundError:
            print("✗ unzip not found. Trying Python zipfile method...")
            try:
                with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                    zip_ref.extractall(data_dir)
                    print("✓ Extraction complete using Python!")
            except Exception as e:
                print(f"✗ Extraction failed: {e}")
                print("  Please install unzip or run manually:")
                print("  unzip KuaiRec.zip")
else:
    print("\n⚠ Zip file not found. Please check the download step above.")

# Return to original directory
os.chdir(original_cwd)
```

## Step 3: Process and Prepare Dataset


```python
# Check if files exist
big_matrix_path = data_dir / 'big_matrix.csv'
small_matrix_path = data_dir / 'small_matrix.csv'

# Also check in subdirectories (sometimes zip extracts to a subfolder)
if not big_matrix_path.exists():
    # Look for CSV files in subdirectories
    for subdir in data_dir.iterdir():
        if subdir.is_dir():
            potential_big = subdir / 'big_matrix.csv'
            potential_small = subdir / 'small_matrix.csv'
            if potential_big.exists() and not big_matrix_path.exists():
                shutil.move(str(potential_big), str(big_matrix_path))
                print(f"✓ Found and moved big_matrix.csv from {subdir.name}")
            if potential_small.exists() and not small_matrix_path.exists():
                shutil.move(str(potential_small), str(small_matrix_path))
                print(f"✓ Found and moved small_matrix.csv from {subdir.name}")

if big_matrix_path.exists():
    print("✓ Found big_matrix.csv")
    # Load and check format
    sample_df = pd.read_csv(big_matrix_path, nrows=5)
    print(f"\nFile columns: {sample_df.columns.tolist()}")
    print(f"Sample data:")
    print(sample_df.head())
    
    # Check if we need to rename columns
    if 'user_id' in sample_df.columns and 'video_id' in sample_df.columns:
        print("\n✓ File has user_id and video_id columns")
        print("  We'll rename them to userId and itemId for consistency")
    elif 'userId' in sample_df.columns and 'itemId' in sample_df.columns:
        print("\n✓ File already has userId and itemId columns")
    else:
        print("\n⚠ Warning: Unexpected column names. Please check the file format.")
else:
    print("✗ big_matrix.csv not found")
    print(f"  Please check the download and extraction steps above")
    print(f"  Expected location: {big_matrix_path}")

if small_matrix_path.exists():
    print("\n✓ Found small_matrix.csv (test subset)")
else:
    print("\nℹ small_matrix.csv not found (optional, can use big_matrix.csv for both train/test)")
```

```python
# Load and prepare KuaiRec dataset
if big_matrix_path.exists():
    print("Loading KuaiRec dataset...")
    print("="*60)
    
    # Load big_matrix.csv
    print("Loading big_matrix.csv...")
    ratings = pd.read_csv(big_matrix_path)
    
    # Rename columns to standard format if needed
    column_mapping = {
        'user_id': 'userId',
        'video_id': 'itemId',
        'watch_ratio': 'rating'  # Use watch_ratio as rating
    }
    
    for old_col, new_col in column_mapping.items():
        if old_col in ratings.columns:
            ratings[new_col] = ratings[old_col]
            if old_col != new_col:
                ratings = ratings.drop(columns=[old_col])
    
    # If rating column doesn't exist, create it from watch_ratio or use a default
    if 'rating' not in ratings.columns:
        if 'watch_ratio' in ratings.columns:
            ratings['rating'] = ratings['watch_ratio']
        else:
            print("⚠ Warning: No rating/watch_ratio column found. Using default rating=1")
            ratings['rating'] = 1.0
    
    # Keep only necessary columns
    required_cols = ['userId', 'itemId', 'rating']
    ratings = ratings[[col for col in required_cols if col in ratings.columns]]
    
    print(f"\nDataset loaded successfully!")
    print(f"Total interactions: {len(ratings)}")
    print(f"Unique users: {ratings['userId'].nunique()}")
    print(f"Unique items: {ratings['itemId'].nunique()}")
    print(f"\nRating statistics:")
    print(ratings['rating'].describe())
    
    # Save as combined CSV for experiments
    combined_csv = data_dir / 'kuairec_combined.csv'
    ratings.to_csv(combined_csv, index=False)
    print(f"\n✓ Saved combined dataset to: {combined_csv}")
    
    # Also save train/test split if small_matrix exists
    if small_matrix_path.exists():
        print("\nLoading small_matrix.csv (test set)...")
        test_ratings = pd.read_csv(small_matrix_path)
        
        # Apply same column mapping
        for old_col, new_col in column_mapping.items():
            if old_col in test_ratings.columns:
                test_ratings[new_col] = test_ratings[old_col]
                if old_col != new_col:
                    test_ratings = test_ratings.drop(columns=[old_col])
        
        if 'rating' not in test_ratings.columns:
            if 'watch_ratio' in test_ratings.columns:
                test_ratings['rating'] = test_ratings['watch_ratio']
            else:
                test_ratings['rating'] = 1.0
        
        test_ratings = test_ratings[[col for col in required_cols if col in test_ratings.columns]]
        
        # Save test set
        test_csv = data_dir / 'test.csv'
        test_ratings.to_csv(test_csv, index=False)
        print(f"✓ Saved test set to: {test_csv}")
        print(f"  Test interactions: {len(test_ratings)}")
        
        # Train set = big_matrix - test set
        train_ratings = ratings[~ratings.set_index(['userId', 'itemId']).index.isin(
            test_ratings.set_index(['userId', 'itemId']).index)]
        train_csv = data_dir / 'train.csv'
        train_ratings.to_csv(train_csv, index=False)
        print(f"✓ Saved train set to: {train_csv}")
        print(f"  Train interactions: {len(train_ratings)}")
    else:
        print("\nℹ Using big_matrix.csv for both train and test (will split in notebooks)")
    
    print("\n" + "="*60)
    print("Dataset preparation complete!")
    print("="*60)
else:
    print("\n⚠ Please download big_matrix.csv first from https://kuairec.com/")
```

```python
# Display sample data and statistics
if big_matrix_path.exists():
    print("\nSample Data:")
    print(ratings.head(10))
    
    print("\n" + "="*60)
    print("Dataset Statistics:")
    print("="*60)
    print(f"Total interactions: {len(ratings):,}")
    print(f"Unique users: {ratings['userId'].nunique():,}")
    print(f"Unique items: {ratings['itemId'].nunique():,}")
    print(f"\nRating distribution:")
    print(ratings['rating'].value_counts().sort_index().head(20))
    
    # Interactions per user
    user_counts = ratings.groupby('userId').size()
    print(f"\nInteractions per user:")
    print(f"  Mean: {user_counts.mean():.2f}")
    print(f"  Min: {user_counts.min()}")
    print(f"  Max: {user_counts.max()}")
    
    # Interactions per item
    item_counts = ratings.groupby('itemId').size()
    print(f"\nInteractions per item:")
    print(f"  Mean: {item_counts.mean():.2f}")
    print(f"  Min: {item_counts.min()}")
    print(f"  Max: {item_counts.max()}")
```
