# Fill Unbiased Metrics

Use this notebook to collect **unbiased** evaluation metrics and write **`data/unbiased_metrics.csv`**.

**Where to get the numbers:**
1. Open the corresponding **biased** notebook (see STEPS.md).
2. Load the **unbiased** data instead of the biased data:
   - **YAHOO:** `../../data/yahoo_data/random.txt`
   - **COAT:** `../../data/coat_data/coat_data/coat/test.csv`
   - **KUAIREC:** `../../data/kuairec_data/test.csv` (or your full-feedback protocol)
3. Use the **same** model (load the saved .keras), same preprocessing (user/item encoding), and evaluate on the unbiased test set.
4. Compute RMSE, MAE, NDCG@5, NDCG@10, Recall@5, Recall@10 (same as in the biased notebook).
5. Paste the six numbers into the dict below for that (Dataset, Architecture).

Then run the **Save to CSV** cell and re-run **Bias_Magnitude_Analysis.ipynb**.

```python
import os
import numpy as np
import pandas as pd

ROOT = os.getcwd()
DATA_DIR = os.path.join(ROOT, "data")
OUT_CSV = os.path.join(DATA_DIR, "unbiased_metrics.csv")
COLS = ["RMSE", "MAE", "NDCG@5", "NDCG@10", "Recall@5", "Recall@10"]
```

## Option 1: Load existing CSV and edit (if you already have partial data)

Run this cell to load `data/unbiased_metrics.csv` (or the template). Edit `df_edit` in the next cell and then run **Save to CSV**.

```python
template = os.path.join(DATA_DIR, "unbiased_metrics_template.csv")
path = OUT_CSV if os.path.isfile(OUT_CSV) else template
df_edit = pd.read_csv(path)
print("Loaded:", path)
df_edit
```

## Option 2: Fill one row at a time (paste your 6 metrics)

Replace the `None` values with your numbers, then run the cell. Run this cell **once per (Dataset, Architecture)** you want to update; then run **Save to CSV**.

```python
# Paste your unbiased metrics here. Example for YAHOO-MF:
# RMSE=0.32, MAE=0.26, NDCG@5=0.80, NDCG@10=0.82, Recall@5=0.78, Recall@10=0.85

DATASET = "YAHOO"   # or "COAT" or "KUAIREC"
ARCH    = "MF"     # or "NCF" or "LightGCN"
RMSE    = None     # float
MAE     = None     # float
NDCG5   = None     # float
NDCG10  = None     # float
Recall5 = None     # float
Recall10= None     # float

if DATASET and ARCH and RMSE is not None:
    if 'df_edit' not in dir() or df_edit is None:
        path = OUT_CSV if os.path.isfile(OUT_CSV) else template
        df_edit = pd.read_csv(path)
    mask = (df_edit["Dataset"] == DATASET) & (df_edit["Architecture"] == ARCH)
    if mask.any():
        df_edit.loc[mask, "RMSE"] = RMSE
        df_edit.loc[mask, "MAE"] = MAE
        df_edit.loc[mask, "NDCG@5"] = NDCG5
        df_edit.loc[mask, "NDCG@10"] = NDCG10
        df_edit.loc[mask, "Recall@5"] = Recall5
        df_edit.loc[mask, "Recall@10"] = Recall10
        print(f"Updated {DATASET} {ARCH}")
    else:
        new_row = {"Dataset": DATASET, "Architecture": ARCH, "Model": "Naive ERM",
                   "RMSE": RMSE, "MAE": MAE, "NDCG@5": NDCG5, "NDCG@10": NDCG10,
                   "Recall@5": Recall5, "Recall@10": Recall10}
        df_edit = pd.concat([df_edit, pd.DataFrame([new_row])], ignore_index=True)
        print(f"Appended {DATASET} {ARCH}")
    display(df_edit)
else:
    print("Set DATASET, ARCH, and at least RMSE (and other metrics), then re-run.")
```

## Build from a single dict (all metrics at once)

Fill the dict below with your unbiased metrics. Use `None` or `np.nan` for missing values. Then run this cell and the **Save to CSV** cell.

```python
# One entry per (Dataset, Architecture). Each value: (RMSE, MAE, NDCG@5, NDCG@10, Recall@5, Recall@10)
unbiased_results = {
    "YAHOO-MF":       (None, None, None, None, None, None),
    "YAHOO-NCF":      (None, None, None, None, None, None),
    "YAHOO-LightGCN": (None, None, None, None, None, None),
    "COAT-MF":        (None, None, None, None, None, None),
    "COAT-NCF":       (None, None, None, None, None, None),
    "COAT-LightGCN":  (None, None, None, None, None, None),
    "KUAIREC-MF":     (None, None, None, None, None, None),
    "KUAIREC-NCF":    (None, None, None, None, None, None),
    "KUAIREC-LightGCN": (None, None, None, None, None, None),
}

rows = []
for key, vals in unbiased_results.items():
    d, a = key.split("-")
    row = {"Dataset": d, "Architecture": a, "Model": "Naive ERM"}
    for i, c in enumerate(COLS):
        row[c] = vals[i] if i < len(vals) and vals[i] is not None else np.nan
    rows.append(row)
df_edit = pd.DataFrame(rows)
print("Built DataFrame from unbiased_results:")
display(df_edit)
```

## Save to CSV

Writes **`data/unbiased_metrics.csv`**. Then open **Bias_Magnitude_Analysis.ipynb** and run all cells.

```python
os.makedirs(DATA_DIR, exist_ok=True)
df_edit.to_csv(OUT_CSV, index=False)
print("Saved:", OUT_CSV)
print("Next: run Bias_Magnitude_Analysis.ipynb (Run All).")
```
